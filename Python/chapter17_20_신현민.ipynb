문제1  

print_even 리스트 내에 있는 값 중 짝수만 화면에 출력되도록 print_score 함수를 생성 후 결과값을 출력해주세요

print_even ([1, 3, 2, 10, 12, 11, 15])

```python
def print_score(source):
    #반환용 리스트 생성
    answer = []
    #반복하며 짝수 찾기
    for i in range(len(source)):
        if source[i] % 2 == 0:
            answer.append(source[i])
    return answer

print_even = [1, 3, 2, 10, 12, 11, 15]

print(print_score(print_even))
```

뮨제7  

오즈 (oz) 클래스에서 이름, 나이, 성별을 출력하는 who() 메소드를 추가하고 who()메소드를 이용해 아래 출력 결과를 만들어주세요


출력 결과 : 이름: 김코, 나이: 30, 코딩레벨: A

```python
class Oz:
    name = ""
    age = 0
    level = ""
    def __init__(self, name, age, level):
        self.name = name
        self.age = age
        self.level = level
    def who(self):
        print(f'이름 : {self.name}, 나이 : {self.age}, 코딩레벨 : {self.level}')

coding = Oz('신현민', 43, 'A')
coding.who()
```
