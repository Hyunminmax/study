## 문제11

이름 : 나이썬, 김자바, 김씨, 박씨샵

````python
#등급에 , 표기가 있는 경우
이름 = "나이썬", "김자바", "김씨", "박씨샵"
등급 = "B", "B", "A", "C"

students = dict(zip(이름, 등급))
print(f'등급이 , 표기되어 입력된 경우 : {students}')



#등급에 , 표기가 없는 경우
이름1 = "나이썬", "김자바", "김씨", "박씨샵"
등급1 = "B B A C"

등급_str = list(등급1.split())
students1 = dict(zip(이름1, 등급_str))
print(f'등급이 , 표기없이 입력된 경우 : {students1}')

````
