```python
#붕어빵 3종(팥, 슈크림, 잡채)
#각각 가격(2000,2500,3000)
#붕어빵(잡채)3개, 붕어빵(슈크림)2개 주문하면
#"주문이 완료되었습니다. 붕어빵(슈크림) 2개, 붕어빵(잡채) 3개, 총 5개 결제 금액은 14000원 입니다."

#붕어빵 종류
fish__been = "붕어빵(팥)"
fish__cream = "붕어빵(슈크림)"
fish__noodle = "붕어빵(잡채)"

#붕어빵 종류별 가격
fish__been__price = 2000
fish__cream__price = 2500
fish__noodle__price = 3000

#붕어빵 종류별 카운트
fish__been__count = 0
fish__cream__count = 2
fish__noodle__count = 3

#총 주문량과 가격
order__total = fish__cream__count+fish__noodle__count
order__price = fish__cream__price*fish__cream__count+fish__noodle__price*fish__noodle__count



#주문 내역
print(f'주문이 완료되었습니다. {fish__cream}{fish__cream__count}개, {fish__noodle}{fish__noodle__count}개 총 {order__total}개 결제 금액은 {order__price}원 입니다.')
```
