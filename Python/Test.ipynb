```python
#ohlc 리스트에는 시가(open), 고가 (high), 저가 (low) , 종가(close)가 날짜별로 저장돼 있다.
#종가가 150원보다 큰 경우에만 종가를 출력하는 코드를 작성해주세요

ohlc = [["open", "high", "low", "close"], [100, 110, 70, 100], [200, 210, 180, 190], [300, 310, 300, 310]]

for i in range(1,len(ohlc)):
    if ohlc[i][len(ohlc[i])-1] > 150:
        print(ohlc[i][len(ohlc[i])-1])
```

```python
#while문을 사용하여, 100 이상의 자연수 중 가장 작은 23의 배수를 츨력하는 코드를 작성해주세요
x = 100
while x >= 100:
    if x % 23 == 0:
        print(x)
        break
    x += 1
```
