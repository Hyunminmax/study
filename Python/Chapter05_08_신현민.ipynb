##문제2

변수명 전화번호에 들어 있는 번호를 모두 붙여서 출력되도록 코드를 작성해주세요

전화번호 = "010-1234-5678"

출력 예시 : "01012345678"

```python
# 변수 설정
전화번호 = "010-1234-5678"

#####1번#####
# 출력문
print("첫번째 방법")
print(f'전화번호 = {전화번호}')
print(f'{전화번호[0:3]}{전화번호[4:8]}{전화번호[9:]}\n')


#####2번#####
# 변수 설정
전화번호 = "010-123-5678"

# 전화번호 국번 추출
tel_1_text = 전화번호[:전화번호.index('-')]

# 전화번호 국번 끝 인덱스
tel_1_index = int(전화번호.index('-'))

# 1차 추출된 전화번호
tel_value = tel_1_text

# 추출되지 않은 전화번호 입력
temp_text = 전화번호[tel_1_index+1:]

# 전화번호 앞자리 추출
tel_2_text = temp_text[:temp_text.index('-')]

# 전화번호 앞자리의 끝 인덱스
tel_2_index = int(temp_text.index('-'))

# 2차 추출된 전화번호 병합
tel_value += tel_2_text

# 전화번호 뒷자리 추출
tel_3_text = temp_text[temp_text.index('-')+1:]

# 3차 추출된 전화번호 병합
tel_value += tel_3_text
print("두번째 방법: 변수의 자리수 변화에 대응")
print(f'전화번호 = {전화번호}')
print(tel_value+'\n')

#####3번#####
# 변수 설정
전화번호 = "010-123-58"

# 전화번호 국번 추출
tel_text = 전화번호[:전화번호.index('-')]

# 전화번호 국번 끝 인덱스
tel_index = int(전화번호.index('-'))

# 1차 추출된 전화번호
tel_value = tel_text

# 추출되지 않은 전화번호 입력
temp_text = 전화번호[tel_index+1:]

# 전화번호 앞자리 추출
tel_value += temp_text[:temp_text.index('-')]

# 전화번호 앞자리의 끝 인덱스
tel_index = int(temp_text.index('-'))

# 전화번호 뒷자리 추출
tel_value += temp_text[temp_text.index('-')+1:]

print("세번째 방법: 변수의 자리수 변화에 대응")
print(f'전화번호 = {전화번호}')
print(tel_value)
```

##문제3

창의성, 시장성, 안정성, 수익성에 대한 각각의 평가 중 아래 기준에 부합하지 못하면 창업지원을 받을 수 없습니다.

- 수익성 60점 이상
- 창의성과 시장성 50점 이상
- 시장성과 수익성을 더한 값의 절반이 안정성 보다 높아야 함

다음 점수가 입력되었을때 창업 지원이 가능한지 여부를 True 또는 False로 출력해주세요

---

창의성 : 62
시장성 : 71
안정성 : 68
수익성 : 66

```python
# 평가점수
창의성 = 62
시장성 = 71
안정성 = 68
수익성 = 66

# 평가결과
result = 수익성 >= 60 and (창의성 >= 50 and 시장성 >= 50) and ((시장성 + 수익성)/2 > 안정성)

# 평가결과 출력
print(f'평가기준 점수 이상 여부 : {result}')
print(result and "지원하신 기업은 창업지원이 가능합니다.")
print(result == False and "지원하신 기업은 창업지원이 불가능합니다.")
```
