```python
class 빵:
    재료 = "밀가루"

class 붕어빵(빵):
    모양 = "붕어"
    주재료 = ""
    이름 = ""
    def __init__(self, fill):
        self.주재료 = fill
        self.이름 = fill + "붕어빵"

슈 = 붕어빵("슈크림")
print(슈.재료)
print(슈.이름)
```

```python
class A:
  def foo(self):
    print('A.foo() called')


class B(A):
  def foo(self):
    print('B.foo() called')
    super().foo()


class C(A):
  def foo(self):
    print('C.foo() called')
    super().foo()


class D(B, C):
  pass


D().foo()
```
