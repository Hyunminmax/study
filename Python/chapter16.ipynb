```python
# 초기코드
#fish__p, fish__s, fish__j = "붕어빵(팥)", "붕어빵(슈크림)", "붕어빵(잡채)"
#fish__p__sell, fish__s__sell, fish__j__sell = 2000, 2500, 3000
#order_s, order_j = 2, 3
#order_s__pay, order_j__pay = fish__s__sell * 2, fish__j__sell * 3
#total = order_s + order_j
#total_pay = osrder_s__pay + order_j__pay

#print(f'주문이 완료되었습니다. {fish__s} {order_s}개, {fish__j}개 {order_j}총 {total}개 결제 금액은 {total_pay}원 입니다.')


#리팩토링

#메뉴설정
fish = {"팥":2000, "슈크림":2500, "잡채":3000}
fish_menu = list(fish)

#인사말 출력
print("안녕하세요. 붕어빵 인터네셔널입니다.")
print("항상 방금나온 따뜻한 붕어빵을 드립니다. \n아래 메뉴에서 구매하실 붕어빵의 종류와 수량을 입력해주세요.\n")
print("#####  붕어빵 메뉴  #####")

#메뉴 출력
for i in range(len(fish_menu)):
    print(f'{fish_menu[i]} : {fish[fish_menu[i]]}원')

#주문 예시
print("\n주문예시1")
print("팥맛 붕어빵 3개, 슈크림맛 붕어빵 2개, 잡채맛 붕어빵, 2개를 구매하는 경우")
print("팥, 3, 슈크림, 2, 잡채, 2\n")
print("주문예시2")
print("팥맛 붕어빵 3개, 잡채맛 붕어빵, 2개를 구매하는 경우")
print("팥, 3, 잡채, 2")
print("주문은 위와 같이 입력해주세요.\n")

#주문 입력
order = input("따끈따끈한 붕어빵 얼마나 드릴까요?")
order_list = tuple(order.replace(" ", "").split(","))

#주문 검증용
order_confirm = {}

#주문 검증
for i in range(0,len(order_list),2):
    key = order_list[i]
    value = int(order_list[i+1])
    order_confirm[key] = value

#메뉴에 없는 주문 제거 ##########   오동작중
#오동작 원인은 인풋값 order를 split 했을때 공백이 포함되어 들어옴.
#오동작 해결 replace를 통해 공백 제거
order_check = list(set(fish.keys()) & set(order_confirm.keys()))

#주문 확정
for i in order_check:
    if i == "팥":
        bean_price = int(order_confirm[i]) * int(fish[i])
    elif i == "슈크림":
        cream_price = int(order_confirm[i]) * int(fish[i])
    else:
        noodle_price = int(order_confirm[i]) * int(fish[i])

#주문 확인 및 결제 요청
print('주문 하신 내용은 아래와 같습니다.\n')
total_price = 0

for i in range(len(order_check)):
    print(f'{order_check[i]} 붕어빵 ', end="")
    print(f'{fish[order_check[i]]}원 X ', end="")
    print(f'{order_confirm[order_check[i]]}개 = ', end="")
    print(fish[order_check[i]] * order_confirm[order_check[i]])
    total_price += fish[order_check[i]] * order_confirm[order_check[i]]
print(f'\n총 결제 금액은 {total_price}원 입니다.')

#미해결 케이스
#메뉴만 입력하고 수량 입력 안하는 경우
#수량만 입력하고 메뉴를 입력 안하는 경우
#등등
```
