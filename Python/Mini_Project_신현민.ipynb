#### 과제 설명

이번 개발 주제는 '행맨' 게임이구요.

행맨 게임은 아주 유명한 단어 퀴즈 프로그램이예요

어떤 단어가 주어지면 그 단어의 길이만큼 빈 칸(밑줄) 이 주어지고
플레이어는 알파벳을 하나씩 얘기하면서 단어를 추측해요!

10번만에 맞히지 못하면 게임이 끝납니다. 🥲



#### 예시를 들어볼께요!!

만약 E 가 단어에 포함돼있다면, 해당하는 위치에 E 를 표시해줍니다

_ _ _ _ E E
이렇게 되겠네요.

이번에는 A를 해볼께요
A 는 포함되지 않으므로 기회가 줄어들고 지금까지 맞힌 내용 그대로 출력합니다.

_ _ _ _ E E
변화가 없지요.

이번에는 C 를 해봅시다.

그럼
C _ _ _ E E

다음에는 F 를
C _ F F E E

마지막으로 O 를 해보면
C O F F E E 로 게임은 성공으로 끝나게 됩니다.

##본격적인 행맨 게임 규칙 설명

- words 리스트에 추가된 단어들을 사용해주세요
  - words = ["apple", "banana", "orange", "grape", "lemon"]

- 단어의 길이에 맞게 밑줄을 출력해주세요  
  (예)banana의 경우 _ _ _ _ _ _

- 사용자부터 1글자씩 입력을 받되,
   - 단어에 입력값이 포함되면 "단어를 맞추셨군요. 더 힘내봐요"라고 출력해주세요"
   - 단어에 입력값이 포함되어 있지 않다면 "틀렸습니다. 남은 시도 횟수 "란 메시지와 함꼐 남은 횟수를 출력해주세요

- 매번 입력을 받을 때마다 현재까지 맞힌 글자들 표시(맞히면 글자는 밑줄 출력)
    - 예) a 입력 시 : a _ _ _ _
          p 입력 시 : a p p _ _
          c 입력 시 : a p p _ _

- 모든 단어를 맞췄을 경우 "축하합니다! 단어를 맞추셨습니다"와 시도 횟수를 출력해주세요


행맨은 정말 유명 게임으로 찾아볼 수 있는 자료들이 많습니다.   
문제가 어려우신 경우 구글링 통해 문제를 해결해보세요.  

```python
#Hangman게임 만들기

import random

#문제 Pool
words = ["apple", "banana", "orange", "grape", "lemon"]
#문제표기
word_q =""
#입력값의 입력 기회는 10회
input_count = 10

#####문제 출제#####
def start():
    #전역변수 호출
    global word_q
    global input_count
    global words
    global word_a

    #랜덤 선택
    word_a = words[random.randint(0, len(words)-1)].upper()
    #선택된 문제값의 알파벳 길이 만큼 _ 표시
    word_q ="_" * len(word_a)
    print(f'자~ 오늘도 어김없이 찾아온 행맨! 입니다. \n기회는 딱! 10번 당신은 살아남을수 있을까?!\n오늘의 문제는: {word_q}')

    #유저 입력 호출
    while input_count > 0:
        user_try()
#####문제 출제#####


#####유저의 입력#####
def user_try():
    #전역변수 호출
    global input_count
    #문제와 대조할 입력값
    input_try = input()
    #입력 검증 호출
    #입력 기회 1회 차감
    input_count -= 1
    check(input_try.upper(), word_a)

#####유저의 입력#####




#####입력 검증#####
def check(input_try, word_a):
    #사용자 입력값 할당
    input_try = input_try
    word_a = word_a
    #전역변수 호출
    global word_q
    global input_count
    pic = ['Σ(￣□￣;)', '( Ĭ ^ Ĭ ) ', 'ಡ︷ಡ)', '(•ε•；)', '(ﾟヘﾟ)？', '( •́ ̯•̀ )', '(ง •̀_•́)ง', '٩(๑•̀o•́๑)و', '๑و•̀Δ•́)و', "٩( 'ω' )و"]
    #유저의 입력값과 문제값의 모든 인덱스와 비교 후 같은 값이 있으면 change()호출, 없으면 꽝 반환
    if input_try in word_a:
        change(word_a, input_try, pic, input_count)
    else:
        print(f'{input_try}는 없습니다.')
        if input_count != 0:
            print(f'틀렸습니다. {pic[input_count]}\n남은 기회는 {input_count}회 입니다.')
            print(f'오늘의 문제는: {word_q}')
        else:
            print(f'기회를 모두 소진했습니다. {pic[input_count]}\np다시 도전하려면 Coin을 넣어주세요.')
    # 입력값과 문제값의 모든 인덱스와 비교 후 같은 값이 없으면 "틀렸습니다. 남은 시도 횟수"와 남은 횟수(입력 기회 - 입력값을 받은 횟수) 출력
#####입력 검증#####


#####밑줄 변경#####
def change(word_a, input_try, pic, input_count):
    #사용자 입력값 할당
    word_a = word_a
    input_try = input_try
    pic = pic
    input_count = input_count
    #전역변수 호출
    global word_q
    word_q = list(word_q)

    #유저의 입력값과 문제값의 모든 인덱스와 비교 후 같은 값이 있으면 _를 입력값으로 대치
    for i in range(len(list(word_a))):
        if word_a[i] == input_try:
            word_q[i]= input_try
    word_q = ''.join(word_q)

    #맞췄고 기회가 남았다면
    if "_" not in word_q:
        win(word_q)
    else:
        if input_count != 0:
            print('단어를 맞추셨군요. 더 힘내봐요')
            print(f'기회는 {input_count}번 남았습니다. {pic[input_count]}')
            print(f'오늘의 문제는: {word_q}')
        else:
            print(f'기회를 모두 소진했습니다. {pic[input_count]}\n다시 도전하려면 Coin을 넣어주세요.')
#####밑줄 변경#####






#####문제 결과#####
def win(word_q):
    global input_count
    print(f'축하합니다! 단어를 맞추셨습니다.\n{10 - input_count}번만에 맞추다니 대단해요!!!')
    print(f'정답은 : {word_q.upper()}')
    input_count = 0
#####문제 결과#####

start()

```
